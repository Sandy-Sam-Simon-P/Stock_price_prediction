# **Data Preprocessing ðŸ§°**

1. Extracting the features (dependent variable) and target variable (independent variables) from the dataset.

```python
#Feature Extraction
import numpy as np

independent_variable = np.array(df.drop(["Volume","Adj Close","Close","Date"],axis = 1).values)
dependent_variable = np.array(df["Close"].values)
print(independent_variable.shape)
print(dependent_variable.shape)


```

We extracted the features and the target variable, as you can see in the code above. In this case, the features that we will use to predict the close variable are the Open, High, and Low.

2. Splitting the dataset into train and test set to train and validate the regression model.

```python
#Spliting the dataset into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent_variable, dependent_variable, test_size=0.2, random_state=42)
```

# **Training the Regression model ðŸ’ª**

```python
#Training the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
# Making predictions using the test set
pred = model.predict(X_test)
print(pred)
```

We have successfully constructed our regression model utilizing the train data and predicted the target variable "Close" by utilizing the test set.

# **Model Validation âœ…**

1. Finding the Mean squared error, Mean absolute error and r2 score. here the r2 score is considered as the accuracy of the model.

```python
#Validting the trained model using the Mean Squared error
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score
mse = mean_squared_error(y_test,pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, pred)
r2 = r2_score(y_test, pred)

# Print the metrics
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"Mean Absolute Error: {mae}")
print(f"R-squared: {r2}")
```

As shown in the result. With a 99% accuracy rate, our model is overfitting; Checking whether a model is overfitting involves monitoring its performance on both the training and validation (or test) datasets.
